```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
headers = ["p_id", "p_title", "vendor_id", "cluster_id", "cluster_title", "cat_id", "cat_title"]
df = pd.read_csv('datasets/pricerunner_aggregate.csv', header=0, names = headers)
df.head()
```

```python
df[df.cluster_id == 1]
```

```python
# Número de clusters.
print("Number of clusters: ", len(set(df.cluster_id)))
# Número de categorias.
print("Number of categs: ", len(set(df.cat_id)))
```

```python
print(df.cat_title.value_counts())
```

```python
df.drop_duplicates(subset=["cat_id","cat_title"])
```

```python
df.cat_title.value_counts().plot(kind="bar", xlabel="Categs.", ylabel="Qtde", grid=True)
```

```python
def plot_cdf():

    fig, axes = plt.subplots()

cats = list(set(df.cat_title))
cats.sort()
values = []
for cat in cats:
    cat_df = df[df.cat_title == cat]
    pus_by_cat = cat_df.cluster_id.value_counts()
    m = np.mean(pus_by_cat.values)
    s = np.std(pus_by_cat.values)
    values.append([cat, m, s])

info = pd.DataFrame(values, columns=["Cat", "Mean", "Std"])
info.head(20)
```

# Analisando a base SKROUTZ

```python
headers = ["p_id", "p_title", "vendor_id", "cluster_id", "cluster_title", "cat_id", "cat_title"]
df = pd.read_csv('datasets/skroutz_aggregate.csv', header=0, names = headers)
df.head()
```

```python
df[df.cluster_id == 1]
```

```python
# Número de clusters.
print("Number of clusters: ", len(set(df.cluster_id)))
# Número de categorias.
print("Number of categs: ", len(set(df.cat_id)))
```

```python
df.cat_title.value_counts().plot(kind="bar", xlabel="Categs.", ylabel="Qtde", grid=True)
```

```python
df[(df.cat_title != "Men's Watches") & (df.cat_title != "Women's Watches")].cat_title.value_counts(normalize=True).plot(kind="bar", xlabel="Categs.", ylabel="Qtde", grid=True)
```
